```python
from sklearn.datasets import load_iris
from sklearn.utils import shuffle
import pandas as pd
from sklearn.metrics import roc_auc_score
```

```python
X,y = load_iris(return_X_y = True)
#Merging input and outputs
data = pd.concat([pd.DataFrame(X),pd.Series(y)],axis=1)
```

```python
data
```

```python
#Naming columns
data.columns = ['F1','F2','F3','F4','TARGET']
```

```python
data
```

```python
#Shuffling data and resetting index
data = shuffle(data)
data = data.reset_index(drop=True)

#Separating input and output
X = data.iloc[:,:4]
y = data.iloc[:,4]
```

```python
#Train data
X_train = X[:90]
y_train = y[:90]
#Cross Validation data
X_cv = X[90:120]
y_cv = y[90:120]
#Test data
X_test = X[120:]
y_test = y[120:]
```

```python
#Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_cv = sc_X.transform(X_cv)
X_test = sc_X.transform(X_test)
```

```python
#converting y to one hot formats
y_train = pd.get_dummies(y_train).values
y_cv = pd.get_dummies(y_cv).values
y_test = pd.get_dummies(y_test).values
```
