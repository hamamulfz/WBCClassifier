```python
import pandas as pd
import numpy as np
TrainX = pd.read_csv('Xtrain29.csv', header=None)
Trainy = pd.read_csv('Ytrain29.csv', header=None)
TestX = pd.read_csv('Xtest29.csv', header=None)
Testy = pd.read_csv('Ytest29.csv', header=None)
```

```python
import numpy as np
#dataset = np.array(df)
X_train=np.array(TrainX)
y_train=np.array(Trainy)
X_test=np.array(TestX)
y_test=np.array(Testy)
```

```python

y_test=np.ravel(y_test)

y_train=np.ravel(y_train)
```

```python
# Penyekalaan fitur dilakukan pada data latih dan uji.
# Detail kenapa ini penting dijelaskan di sini: http://scikit-learn.org/stable/auto_examples/preprocessing/plot_scaling_importance.html
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

# Fungsi fit_transform hanya dilakukan di data latih
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

#simpan data scaling ke variabel
X_train_scaling = X_train
```

```python

from sklearn.neural_network import MLPClassifier
i=10
data = [ ]
loss = [ ]
iterasi = [ ]
cv_score = [ ]
std_cv = [ ]
train = [ ]
layers = [ ]
while (i<71):
    clf = MLPClassifier(activation='relu',solver='lbfgs', alpha=1e-5,max_iter=2000,
                    hidden_layer_sizes=(i), random_state=1)
    print(clf.fit(X_train, y_train))

    print('Loss Pelatihan : ', clf.loss_)    
    print('Jumlah iterasi : ', clf.n_iter_)
    
    from sklearn.model_selection import cross_val_score
    scoring='accuracy'
    accuracies = cross_val_score(estimator=clf, X=X_train, y=y_train, cv=5,scoring=scoring)
    avg_accuracy = accuracies.mean()
    std_accuracy = accuracies.std()
    print('Akurasi: ', avg_accuracy)
    print('SD: ', std_accuracy)
    
    #Prediksi Data Uji
    result = clf.predict(X_test)

    #Confusion Matriks Hasil Prediksi dengan Nilai Sebenarnya
    from sklearn.metrics import classification_report,confusion_matrix
    print(confusion_matrix(y_test,result))

    # Cek kemampuan kecerdasan buatan dengan data uji 
    test_accuracy = clf.score(X_test, y_test)
    from sklearn.metrics import accuracy_score
    akurasi = accuracy_score(y_test, result)
    print('Akurasi Data Uji:', test_accuracy , '=', akurasi)

    # Cek kemampuan kecerdasan buatan dengan data uji 
    train_accuracy = clf.score(X_train, y_train)
    print('Akurasi Data Latih:', train_accuracy)  

    data.append( float('{0:.4f}'.format(test_accuracy))*100)
    loss.append(float('{0:.8f}'.format(clf.loss_)))
    iterasi.append(clf.n_iter_)
    cv_score.append( float('{0:.4f}'.format(avg_accuracy))*100)
    std_cv.append( float('{0:.4f}'.format(std_accuracy))*100)
    train.append( float('{0:.4f}'.format(train_accuracy))*100)
    layers.append(i)
    print()
    
    if(i>=30):
        i=i+5
    elif(i<30):  
        i=i+1
```

```python
import itertools
import numpy as np
import matplotlib.pyplot as plt

from sklearn import svm, datasets
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, result)
np.set_printoptions(precision=2)

class_names = ['Basofil', 'Netrofil','Monosit','Limfosit']


# Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=class_names,
                      title='Confusion matrix')
```
