```python
'''Klasifikasi Sel Sarah Putih Muda Menggunakan Multilayer Perceptron
'''

#Authors: Hamamul Fauzi <hamamul.fauzi@gmail.com>
#Final Project Thesis for Graduade in Physics Engineering
#Universitas Gadjah Mada, Yogyakarta
#Indonesia

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


'''
Kode sumber ini menggunakan pustaka Scikit-Learn untuk machine learning 
Untuk menginstall pustaka tersebut dapat menggunakan perintah pip
 >> pip install -U scikit-learn

 
jika menggunakan anaconda, dapat melakukan instalasi melalui conda promp 
>> conda install scikit-learn

untuk menghapus library
>> conda remove scikit-learn

dokumentasi pustaka dapat dilihat pada:
http://www.scikit-yb.org/en/latest/quickstart.html#walkthrough
http://scikit-learn.org/stable/install.html
'''

import itertools
from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report,confusion_matrix

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    Fungsi ini akan mencetak dan menggambarkan confusion matrix
    dipelajari dari http://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html#sphx-glr-auto-examples-model-selection-plot-confusion-matrix-py
    
    Parameters
    ---------------
    cm : confusion matrix, already defined with cnf_matrix = confusion_matrix(y_test, result)
    
    normalize : True or False, opsi untuk melakukan normalisasi atau tidak pada nilai cm
    
    title : String, judul untuk plot window confusion matrix
    
    cmap : warna dari plot cm
    """
    
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

def plot_classification_report (classifier, classes,
                      X_train, y_train,
                      X_test, y_test):
    '''Fungsi ini digunakan untuk menampilkan laporan klasifikasi sebagai evaluasi metrik
    dipelajari dari http://www.scikit-yb.org/en/latest/api/classifier/classification_report.html
    
    Parameter
    -------------
    classifier : metode klasifikasi yang digunakan
    
    classes : String,  label dari target klasifikasi
    
    X_train : array, dataset input pelatihan
    
    y_train : dataset target pelatihan
    
    X_test :  array, dataset input pengujian
    
     y_test : dataset target hasil pengujian
    '''
    visualizer = ClassificationReport(classifier, classes=class_names, support=True)
    visualizer.fit(X_train, y_train)  # Fit the visualizer and the model
    visualizer.score(X_test, y_test)  # Evaluate the model on the test data
    g = visualizer.poof()             # Draw/show/poof the data
```
